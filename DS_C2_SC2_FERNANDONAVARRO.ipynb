## **MANEJO DE ARCHIVO CSV**

```python
import glob, os
import pandas as pd

# ARCHIVO EXISTENTE
def file_exists(name):
    file =  glob.glob(file_name)
    #print(file)
    return len(file) > 0

# ARCHIVO CON DATOS
def file_is_not_empty(name):
    return os.stat(name).st_size > 0

mierror = 0
path = '/content/sample_data'
file_name = path+ '/country_vaccinations.csv'
if file_exists(file_name):
    if file_is_not_empty(file_name):
        file_handler = pd.read_csv(file_name)
    else:
        print('El archivo no tiene información')
        mierror = 1
else:
    print('El archivo no existe')
    mierror = 1
```

###**a. Información del archivo**

```python
# --------------------
# a. INFORMACIÓN
# --------------------
print("a. INFORMACIÓN DEL ARCHIVO")
print('Tamaño del resultado: ', file_handler.shape)
print(file_handler.head(30))
print()
```

###**b. Estructura y tipos de datos**

```python
# --------------------
# b. ESTRUCTURA
# --------------------
print("b. ESTRUCTURA Y TIPOS DE DATOS")
print(file_handler.info())
print()

# Fechas tipo datetime64
print("***Conversión de fechas tipo datetime64***\n")
file_handler['date'] = pd.to_datetime(file_handler['date'])
print(file_handler.dtypes)
print()
```

###**c. Vacunas por compañía**

```python
# ------------------------
# c. VACUNAS POR COMPAÑÍA
# ------------------------
print("c. VACUNAS POR COMPAÑÍA")
vaccinations = file_handler[["vaccines", "daily_vaccinations"]].groupby('vaccines').sum()
print(vaccinations.head(30))
print()
```

###**d. Vacunas totales**

```python
# ------------------------
# d. VACUNAS TOTALES
# ------------------------
print("d. VACUNAS APLICADAS EN TODO EL MUNDO")
totvaccinations = file_handler[["daily_vaccinations"]].sum()
print('Total de vacunas en el mundo: ',  format(totvaccinations["daily_vaccinations"], '0,.0f'))
print()
```

###**e. Promedio (diario) por país**

```python
# ------------------------
# e. PROMEDIO (Entiendo que Diario porque no dice) POR PAÍS
# ------------------------
print("e. PROMEDIO (DIARIO) POR PAÍS")
dailyvaccinations = file_handler[["country", "daily_vaccinations"]].groupby('country').mean()
print(dailyvaccinations)
print()
```

###**f. Vacunas totales del 29/01/21**

```python
# ------------------------
# f. VACUNAS TOTALES DEL 29/01/21
# ------------------------
import datetime
from datetime import date

print("f. VACUNAS DEL 29/01/21 EN TODO EL MUNDO")
mifecha = datetime.datetime(2021, 1, 29)
mydatevaccinations = file_handler.loc[file_handler["date"]==mifecha, "daily_vaccinations"].sum()
print(date.strftime(mifecha,'%b %d, %Y'), ', total de vacunas: ', format(mydatevaccinations, "0,.0f"))
print()
```

###**g. Dataframe con columna derivada de diferencias**

```python
# ------------------------
# g. DATAFRAME CON COLUMNA DERIVADA DIFERENCIAS
# ------------------------
print("g. DATAFRAME CON COLUMNA DE DIFERENCIAS DE FECHAS")
conDiferencias = file_handler.copy()
conDiferencias['diferencias'] = conDiferencias["daily_vaccinations"] -conDiferencias["daily_vaccinations_raw"]
print(conDiferencias)
print()
```

###**h. Periodo de aplicación registrado**


```python
# ------------------------
# h. PERIODO DE APLICACIÓN
# ------------------------

import datetime
from datetime import date

print("h. PERIODO DE APLICACIÓN DE VACUNAS REGISTRADO")
mifechaini = file_handler["date"].min()
mifechafin = file_handler["date"].max()

periodo = mifechafin - mifechaini
print('Periodo de aplicación de vacunas de', periodo.days, 'días')
print('De', date.strftime(mifechaini,'%b %d, %Y'), 'a', date.strftime(mifechafin,'%b %d, %Y'))
print()
```

###**i. Dataframe con columna derivada de cantidad de tipos de vacunas**

```python
# ------------------------
# i. DATAFRAME CON COLUMNA DERIVADA DE CANTIDAD DE TIPOS DE VACUNAS
# ------------------------
print("i. DATAFRAME CON COLUMNA DE CANTIDAD DE TIPOS DE VACUNAS UTILIZADAS")
conCantidad = file_handler.copy()
#conCantidad["uno"] =  conCantidad["vaccines"].str.len()
#conCantidad["dos"] =  conCantidad["vaccines"].str.replace(',','',).str.len()
conCantidad["canVac"] = conCantidad["vaccines"].str.len()- conCantidad["vaccines"].str.replace(',','',).str.len() + 1
print(conCantidad)
print()
```

###**j. Dataframe con  filtro de fecha**

```python
# ------------------------
# j. DATAFRAME CON FILTRO DE FECHA
# ------------------------
import datetime
from datetime import date
print("j. DATAFRAME CON FILTRO DE FECHA")
mifecha = datetime.datetime(2020, 12, 20)
antes20= file_handler[file_handler["date"] < mifecha]
print('Dataframe con registros de fecha menor a', date.strftime(mifechaini,'%b %d, %Y'))
print('Tamaño del resultado filtrado: ', antes20.shape)
print(antes20.head(30))
print()
```

###**k. Dataframe con  filtro de vacuna**

```python
# ------------------------
# k. DATAFRAME CON FILTRO DE VACUNA
# ------------------------
print("k. DATAFRAME CON FILTRO DE VACUNA")
pfizer= file_handler[file_handler["vaccines"].str.contains('pfizer', case=False)]
print('Dataframe con registros con vacuna Pfizer')
print('Tamaño del resultado filtrado: ', pfizer.shape)
print(pfizer)
print()
```

###**l. Archivo de Excel**

```python
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
